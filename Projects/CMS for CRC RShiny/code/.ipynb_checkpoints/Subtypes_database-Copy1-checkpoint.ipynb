```python
from indra.sources import trips
from indra.statements.concept import Concept
from indra.statements.statements import *
from indra.preassembler import Preassembler
from indra.assemblers.pybel import assembler
from pybel_jupyter import to_jupyter
```

```python
####### NODE OBJECT ##########

#Concept
CMS1 = Concept(name = 'CMS1')
CMS2 = Concept(name = 'CMS2')
CMS3 = Concept(name = 'CMS3')
CMS4 = Concept(name = 'CMS4')
```

```python
#Colorectal cancer disease Agent id
CRC_id = Agent('CRC', db_refs = {'DOID':'DOID:9256'})

CMS1_name = Agent('CMS1_name', db_refs = {'TEXT': 'Immune'})
MSI = Agent('MSI', db_refs = {'TEXT': 'Microsatellite instability'})
CIMP_high = Agent('CIMP_high', db_refs = {'TEXT': 'CIMP HIGH'})
CMS2_name = Agent('CMS2_name', db_refs = {'TEXT': 'Canonical'})
Canonical = Agent('Canonical', db_refs = {'TEXT': 'Canonical'})
#Agent (Proteins, genes)

#Drug id
FOLFOX = {'PUBCHEM': '135659064'}
```

```python
#literature evidence
Fake = Evidence(text = 'fake evidence')
```

```python
#define the relationship between nodes
CMS1_MSI = IncreaseAmount(MSI, CMS1_name,evidence= Fake)
CMS1_CIMP = IncreaseAmount(CIMP_high, CMS1_name, evidence = Fake)
CMS2_Can = IncreaseAmount(CMS2_name, Canonical)
```

```python
#combine all the statements
stmt_list = [CMS1_MSI, CMS1_CIMP, CMS2_Can]
```

```python
pba = assembler.PybelAssembler(stmt_list)
```

```python
#code to give the citation
pba.statements[0]
```

```python
type(pba)
```

```python
#transform the ensemble into pyBEL 
belgraph = pba.make_model()
```
